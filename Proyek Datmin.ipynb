# Evaluasi dan Interpretasi

# Evaluasi 1

# Sum Of Square Errors


```python
from sklearn.cluster import KMeans

x = np.array([[1, 2], [5, 8], [1.5, 1.8], [8, 8], [1, 0.6], [9, 11]])
k_values = range(3, 6)
sse_values = []
```

```python
for k in k_values:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(x)
    sse_values.append(kmeans.inertia_)
```

```python
k_values = [2, 3, 4, 5]
x = np.array([[1, 2], [5, 8], [1.5, 1.8], [8, 8], [1, 0.6], [9, 11]])
sse_scores = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(x)
    sse = kmeans.inertia_ 
    sse_scores.append(sse)
    print(f"For n_clusters = {k}, the SSE is: {sse}")
```

```python
k_values = [2, 3, 4, 5]
sse_values = [10.2, 5.5, 3.8, 2.9]
plt.plot(k_values, sse_values, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Elbow Method for Optimal k')
plt.show()
```

# Evaluasi 2
# Adjusted Rand Score

```python
from sklearn import metrics

x = np.array([[1, 2], [5, 8], [1.5, 1.8], [8, 8], [1, 0.6], [9, 11]])
y_true = np.array([0, 1, 0, 1, 0, 1])
k_values = range(3, 6)
rand_scores = []
```

```python
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(x)
    labels = kmeans.labels_
    rand_score = metrics.adjusted_rand_score(y_true, labels)
    rand_scores.append(rand_score)
    print(f"For n_clusters = {k}, the adjusted Rand score is: {rand_score}")
```

```python
plt.plot(k_values, rand_scores, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Adjusted Rand Score')
plt.title('Elbow Method with Adjusted Rand Score')
plt.show()
```
